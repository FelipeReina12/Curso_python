Aqui implementamos un metodo de extracción de datos por etiqueta mediante loc 

```python
df9.loc["735 Dolores St": "332 Hill St", "Country": "ID"]  #Aqui se selecciona el rango de direcciones primero las columnas y luego las filas
```

```python
#Otro ejemplo a modo de prueba
df9.loc["1056 Sanchez St" : "551 Alvarado St", "Name":"State"]
```

```python
list(df9.loc[:, "Country"])  #Almacenamos los datos de la columna "Country" en una lista
```

```python
#Otro ejemplo
dict(df9.loc[:, "Address"])  #Almacenamos la columna "Address" en un diccionario
#Importante colocar simepre los : antes de la columna para que se tome como un rango de columnas
```

```python
df9
```

Aqui implementamos un nuevo metod de extracción de datos mediante iloc que nos permite acceder a los datos de una matriz de forma más eficiente y flexible.

```python
df9.iloc[1:3, 1:3]  #iloc sirve para buscar datos mediante su posición
```

```python
df9.iloc[2:5, 0:3]  #Para este metodo de busqueda no hay que tomar como referencia los nombres de las columnas ni filas
#Se toman las pocisiones de los datos
```

```python
#Otro ejemplo
df9.iloc[1:, :4]
```

```python
#Otro ejemplo
df9.iloc[3: 5, 4: ]
```

```python
df9
```

```python
df9.iloc[1: 4, 1: 4]
```

Importante para estos métodos de extracción de datos siempre se tienen en cuenta las pocisiones iniciales y finales 
Es decir si quiero un rango desde 1 hasta 8
sería: 1:7


```python
df9.iloc[0:3, 1:3]
```
